```python
# Convert 'Fraud Type' to a binary column (Fraudulent = 1, Legitimate = 0)
df['Is_Fraudulent'] = df['Fraud Type'].notna().astype(int)

# Fraud Cases by Card Type (Bar Chart)
plt.figure(figsize=(10,5))
sns.countplot(x='Card Type', data=df[df['Is_Fraudulent'] == 1], 
              order=df['Card Type'].value_counts().index, palette='Reds')
plt.xticks(rotation=45)
plt.xlabel("Card Type")
plt.ylabel("Fraud Count")
plt.title("Fraud Cases by Card Type")

# Show the plot
plt.show()
```

```python
# Count fraud cases for each merchant type
fraud_by_merchant = df[df['Is_Fraudulent'] == 1]['Merchant'].value_counts()

# Plot the data
plt.figure(figsize=(12, 6))
sns.barplot(x=fraud_by_merchant.index, y=fraud_by_merchant.values, palette="coolwarm")

# Add Title & Labels
plt.title("Fraud Cases by Merchant Type", fontsize=14, fontweight='bold', color='darkred')
plt.xlabel("Merchant Type", fontsize=12, fontweight='bold', color='black')
plt.ylabel("Number of Fraud Cases", fontsize=12, fontweight='bold', color='black')

# Rotate X-axis labels for better visibility
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.show()
```

```python
# Count fraud cases for each location
fraud_by_location = df[df['Is_Fraudulent'] == 1]['Location'].value_counts()

# Generate a color palette based on the number of locations
colors = sns.color_palette("husl", len(fraud_by_location))

# Plot the bar chart
plt.figure(figsize=(12, 6))
plt.bar(fraud_by_location.index, fraud_by_location.values, color=colors, alpha=0.8)

# Add Title & Labels
plt.title("Fraud Cases by Location", fontsize=14, fontweight='bold', color='darkred')
plt.xlabel("Location", fontsize=12, fontweight='bold', color='black')
plt.ylabel("Number of Fraud Cases", fontsize=12, fontweight='bold', color='black')

# Rotate X-axis labels for better visibility
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.show()
```

```python
# Count unique IPs per location
ip_by_location = df.groupby("Location")["IP Address Region"].nunique().reset_index()
ip_by_location = ip_by_location.sort_values(by="IP Address Region", ascending=False)

# Set figure size
plt.figure(figsize=(12, 6))

# Create a bar plot using seaborn
sns.barplot(
    x="Location", 
    y="IP Address Region", 
    data=ip_by_location, 
    palette="viridis"
)

# Add Title & Labels
plt.title("Unique IP Addresses by Location", fontsize=14, fontweight='bold', color='darkblue')
plt.xlabel("Location", fontsize=12, fontweight='bold', color='black')
plt.ylabel("Number of Unique IPs", fontsize=12, fontweight='bold', color='black')

# Rotate X-axis labels for better readability
plt.xticks(rotation=45, ha="right")

# Show the plot
plt.show()
```

```python
# Count occurrences of each age
age_counts = df["User Age"].value_counts().sort_index()

# Set figure size
plt.figure(figsize=(12, 6))

# Create a line plot
sns.lineplot(
    x=age_counts.index,  
    y=age_counts.values,  
    marker="o",
    color="blue"
)

# Add Title & Labels
plt.title("User Count by Age", fontsize=14, fontweight="bold", color="darkblue")
plt.xlabel("User Age", fontsize=12, fontweight="bold", color="black")
plt.ylabel("Number of Users", fontsize=12, fontweight="bold", color="black")

# Show grid for better readability
plt.grid(True, linestyle="--", alpha=0.6)

# Show the plot
plt.show()
```

```python
# Count fraud cases by type
fraud_types = df["Fraud Type"].value_counts()

# Set figure size
plt.figure(figsize=(8, 8))

# Create a pie chart
plt.pie(
    fraud_types, 
    labels=fraud_types.index, 
    autopct='%1.1f%%', 
    colors=['red', 'orange', 'purple', 'blue', 'green'], 
    startangle=140, 
    wedgeprops={'edgecolor': 'black'}
)

# Add Title
plt.title("Distribution of Different Fraud Types", fontsize=14, fontweight="bold", color="darkred")

# Show the chart
plt.show()
```

```python
# Ensure Timestamp is in DateTime format
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')

# Extract Month-Year
df['Month_Year'] = df['Timestamp'].dt.to_period('M').astype(str)

# Group transactions by Month-Year and Device Used
device_trend = df.groupby(["Month_Year", "Device Used"]).size().unstack()

# Set figure size
plt.figure(figsize=(12, 6))

# Create a stacked area plot
device_trend.plot(kind="area", stacked=True, colormap="coolwarm", alpha=0.7, figsize=(12,6))

# Add Title & Labels
plt.title("Device Usage Over Time", fontsize=14, fontweight="bold", color="darkblue")
plt.xlabel("Month-Year", fontsize=12, fontweight="bold", color="black")
plt.ylabel("Number of Transactions", fontsize=12, fontweight="bold", color="black")

# Rotate X-axis labels for readability
plt.xticks(rotation=45, ha="right")

# Show the plot
plt.show()
```

```python
# Count transactions for each Merchant Type and Device Used
device_merchant_counts = df.groupby(["Merchant", "Device Used"]).size().reset_index(name="Transaction_Count")

# Set figure size
plt.figure(figsize=(12, 6))

# Create Bubble Chart
sns.scatterplot(
    x="Merchant", 
    y="Device Used", 
    size="Transaction_Count", 
    hue="Transaction_Count", 
    data=device_merchant_counts, 
    palette="coolwarm",
    sizes=(50, 500), 
    alpha=0.7
)

# Add Title & Labels
plt.title("Bubble Chart: Merchant Type vs. Device Used", fontsize=14, fontweight="bold", color="darkblue")
plt.xlabel("Merchant Type", fontsize=12, fontweight="bold", color="black")
plt.ylabel("Device Used", fontsize=12, fontweight="bold", color="black")

# Rotate X-axis labels for better readability
plt.xticks(rotation=45, ha="right")

# Show the plot
plt.show()
```

```python
# Set figure size
plt.figure(figsize=(10, 6))

# Create a hexbin map
plt.hexbin(df["Longitude"], df["Latitude"], gridsize=30, cmap="Reds", alpha=0.7)

# Add Labels
plt.title("Fraud Density Map", fontsize=14, fontweight="bold", color="darkred")
plt.xlabel("Longitude", fontsize=12, fontweight="bold", color="black")
plt.ylabel("Latitude", fontsize=12, fontweight="bold", color="black")

# Show color bar
plt.colorbar(label="Number of Fraud Cases")

# Show the plot
plt.show()
```

```python
# Count OTP status occurrences
otp_counts = df["OTP Verification Status"].value_counts()

# Set figure size
plt.figure(figsize=(8, 8))

# Create a Pie Chart
plt.pie(
    otp_counts, 
    labels=otp_counts.index, 
    autopct='%1.1f%%', 
    colors=sns.color_palette("coolwarm", len(otp_counts)), 
    startangle=140, 
    wedgeprops={'edgecolor': 'black'}
)

# Add Title
plt.title("Proportion of OTP Verification Status", fontsize=14, fontweight="bold", color="darkred")

# Show the plot
plt.show()
```
