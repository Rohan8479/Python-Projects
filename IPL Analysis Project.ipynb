```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

file_path = r"C:\Users\Rohan\OneDrive\Desktop\Infotact Internship\IPL Database.xlsx"

# Load the "Batting Records" sheet
Batting_Records = pd.read_excel(file_path, sheet_name="Batting Records")

# Display basic info (rows & columns)
print(Batting_Records.info())  

# Display the full dataset in a well-formatted table
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.expand_frame_repr", False)

# Display data
display(Batting_Records)

# Ensure column names are correctly formatted 
Batting_Records.rename(columns=lambda x: x.strip(), inplace=True)

# Aggregate total runs by player
Top_Batsmen = Batting_Records.groupby("Player")["Runs"].sum().reset_index()

# Sort players by total runs
Top_Batsmen = Top_Batsmen.sort_values(by="Runs", ascending=False).head(10)

# Bar Chart: Top 10 Batsmen in IPL History
plt.figure(figsize=(10, 6))

# Create the bar plot with correct order
ax = sns.barplot(data=Top_Batsmen, x="Player", y="Runs", palette="magma", order=Top_Batsmen["Player"])

# Add data labels on top of each bar
for index, row in enumerate(Top_Batsmen.itertuples()):
    plt.text(index, row.Runs + 50, str(row.Runs), ha='center', fontsize=10, color='black', fontweight='bold')

# Formatting the chart
plt.xlabel("Players")
plt.ylabel("Total Runs")
plt.title("Top 10 Batsmen in IPL History")
plt.xticks(rotation=30, ha="right")  
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()  
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

file_path = r"C:\Users\Rohan\OneDrive\Desktop\Infotact Internship\IPL Database.xlsx"

# Load the "Most Six and Four " sheet
Most_Six_and_Four = pd.read_excel(file_path, sheet_name="Most Six and Four ")

# Display basic info (rows & columns)
print(Most_Six_and_Four.info())  

# Display the full dataset in a well-formatted table
pd.set_option("display.max_rows", None)  
pd.set_option("display.max_columns", None)  
pd.set_option("display.expand_frame_repr", False)  

# Display data
display(Most_Six_and_Four) 

# Ensure column names are correctly formatted (strip spaces if necessary)
Most_Six_and_Four.rename(columns=lambda x: x.strip(), inplace=True)

# Sort players by total boundaries (fours + sixes) for ranking
Most_Six_and_Four["Total Boundaries"] = Most_Six_and_Four["4s"] + Most_Six_and_Four["6s"]
Top_Batsmen = Most_Six_and_Four.sort_values(by="Total Boundaries", ascending=False).head(10)

# Bar Chart: Most Fours & Sixes in IPL History
plt.figure(figsize=(12, 6))

# Plot bars for fours and sixes side by side
bar_width = 0.4
players = Top_Batsmen["Player"]
x_indexes = range(len(players))

# Plot Fours
plt.bar(x_indexes, Top_Batsmen["4s"], width=bar_width, label="Fours", color="blue")

# Plot Sixes (shifted slightly for better visibility)
plt.bar([x + bar_width for x in x_indexes], Top_Batsmen["6s"], width=bar_width, label="Sixes", color="red")

# Add data labels
for i, (fours, sixes) in enumerate(zip(Top_Batsmen["4s"], Top_Batsmen["6s"])):
    plt.text(i, fours + 10, str(fours), ha="center", fontsize=10, color="black")
    plt.text(i + bar_width, sixes + 10, str(sixes), ha="center", fontsize=10, color="black")

# Formatting the chart
plt.xlabel("Players")
plt.ylabel("Number of Boundaries")
plt.title("Top 10 Players with Most Fours & Sixes in IPL History")
plt.xticks([x + bar_width / 2 for x in x_indexes], players, rotation=45)
plt.legend()
plt.show()
```
